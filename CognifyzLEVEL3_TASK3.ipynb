```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("/content/drive/Othercomputers/My Laptop/Desktop/Dataset.csv")
```

```python
df.head(2)
```

## Create visualizations to represent the distribution of ratings using different charts (histogram, bar plot, etc.).

```python
plt.figure(figsize=(10, 5))
sns.histplot(df['Aggregate rating'], bins=20, kde=True, color='blue')
plt.xlabel('Ratings')
plt.ylabel('Count')
plt.title('Distribution of Restaurant Ratings')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.countplot(x=df['Aggregate rating'], order=df['Aggregate rating'].value_counts().index, hue=df['Aggregate rating'], palette='viridis')
plt.xlabel('Ratings')
plt.ylabel('Count')
plt.title('Frequency of Each Rating')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.boxplot(x=df['Aggregate rating'], color='red')
plt.xlabel('Ratings')
plt.title('Boxplot of Restaurant Ratings')
plt.show()
```

 ## Compare the average ratings of different cuisines or cities using appropriate visualizations.

```python
top_cuisines = df.groupby('Cuisines')['Aggregate rating'].mean().sort_values(ascending=False).head(10)

plt.figure(figsize=(12, 6))
sns.barplot(x=top_cuisines.values, y=top_cuisines.index, hue=top_cuisines.values, palette="coolwarm")
plt.xlabel('Average Rating')
plt.ylabel('Cuisines')
plt.title('Top 10 Cuisines with Highest Average Ratings')
plt.show()
```

```python
top_cities = df.groupby('City')['Aggregate rating'].mean().sort_values(ascending=False).head(10)

plt.figure(figsize=(12, 6))
sns.barplot(x=top_cities.values, y=top_cities.index, hue=top_cities.values, palette="viridis")
plt.xlabel('Average Rating')
plt.ylabel('City')
plt.title('Top 10 Cities with Highest Average Restaurant Ratings')
plt.show()
```

## Visualize the relationship between various features and the target variable to gain insights.

```python
plt.figure(figsize=(8, 5))
sns.boxplot(x=df['Price range'], y=df['Aggregate rating'], hue=df['Price range'], palette="coolwarm")
plt.xlabel("Price Range")
plt.ylabel("Aggregate Rating")
plt.title("Price Range vs Aggregate Rating")
plt.show()
```

```python
plt.figure(figsize=(6, 5))
sns.boxplot(x=df['Has Online delivery'], y=df['Aggregate rating'], hue=df['Has Online delivery'], palette="viridis")
plt.xlabel("Has Online Delivery")
plt.ylabel("Aggregate Rating")
plt.title("Impact of Online Delivery on Ratings")
plt.show()
```

```python
country_avg_rating = df.groupby('Country Code')['Aggregate rating'].mean().sort_values(ascending=False).head(10)

plt.figure(figsize=(12, 6))
sns.barplot(x=country_avg_rating.index, y=country_avg_rating.values, hue=country_avg_rating.index, palette="magma")
plt.xlabel("Country Code")
plt.ylabel("Average Rating")
plt.title("Top 10 Countries with Highest Average Ratings")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.heatmap(df.select_dtypes(include=['number']).corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Feature Correlation Heatmap")
plt.show()
```

## insights
Higher-priced restaurants tend to have better ratings, possibly due to better service and food quality. Popular restaurants (higher votes) receive better ratings, reinforcing the idea that customer engagement influences perception. Table booking is more common in upscale restaurants, while online delivery is independent of pricing.
